```python
## if value is greater than or equal to 25$ it should return high or else low
```

```python
def billValue(x):
    if (x >= 25):
        return ("high")
    
    elif x >10 :
        return "mid"
    else:
        return ("low")
```

```python
billValue(20)
```

```python
billValue(70)
```

```python
billValue(9)
```

```python
tips["cm_status"] = tips["total_bill"].apply(billValue) ## we are using total bill amount to create new col
```
